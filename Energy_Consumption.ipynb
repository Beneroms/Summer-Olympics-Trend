**Data Analysis and Visualization of 10 Years Hourly Energy Consumption Using Python, Seaborn, And Pandas.**

Step 1:
Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import datetime
%matplotlib inline
```

Step 2:
Read the DataSet

```python
df = pd.read_csv("AEP_hourly.csv")
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

Step 3:
Seperate date and time


```python
df["New_Date"] = pd.to_datetime(df["Datetime"]).dt.date
df["New_Time"] = pd.to_datetime(df["Datetime"]).dt.time
```

```python
df1 = df
df1.head(2)
```

Step 4:

When was the higest Energy Consumption and which Year

Maximum

```python
df1[df1["AEP_MW"] == df["AEP_MW"].max()]
```

Minimum

```python
df1[df1["AEP_MW"] == df["AEP_MW"].min()]
```

Step 5:

Plot and Data visualization

```python
sns.distplot(df1["AEP_MW"])
```

The sns.distplot function has been deprecated in Seaborn and replaced by sns.histplot and sns.displot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df1["AEP_MW"], kde=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(df1["AEP_MW"], kde=True)
plt.show()
```

Step 7:

Extract Date and Time

```python
df1["Year"] = pd.DatetimeIndex(df['New_Date']).year
df1.head(2)
```

Relationship - Energy vs Year

```python
sns.lineplot(x=df1["Year"],y=df1["AEP_MW"], data=df1)
```

Regression

```python
sns.jointplot(x=df1["Year"],
              y=df1["AEP_MW"],
              data=df1,
             kind="reg")
```

```python
sns.jointplot(x=df1["Year"],
              y=df1["AEP_MW"],
              data=df1,
             kind="kde")
```
